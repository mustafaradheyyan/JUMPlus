## 2. Import required libraries

Path of data source

Graphing Y-axis label helper functions

```python
def dec_millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fMM' % (x * 1e-6)

def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.0fMM' % (x * 1e-6)

formatter = ticker.FuncFormatter(millions)
dec_formatter = ticker.FuncFormatter(dec_millions)
```

## 9. Group the sales data of each region on Year_of_Release and plot the same (It should show which region sales was at peak on which year of release)

```python
video_game_sales_sum_by_year = video_game_sales_df.groupby('year_of_release').sum(numeric_only=True)
video_game_sales_sum_by_year.sort_values('games_sold_global', ascending=False)
```

```python
ax = video_game_sales_sum_by_year.plot.line(figsize=(20,8), grid=True)

plt.title('total_video_game_sales_by_year_and_region'.replace('_', ' ').title(), pad=15, fontsize=18)
plt.legend(fontsize = 12)

plt.xlabel("Year of Release", labelpad=10, fontsize=15)
plt.ylabel("Number of Games Sold", labelpad=10, fontsize=15)

ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.yaxis.set_major_locator(ticker.MultipleLocator(50000000))
ax.yaxis.set_major_formatter(formatter)

ax.set_ylim(ymin=0)
ax.set_xlim(xmin=video_game_sales_sum_by_year.index.min(), xmax=video_game_sales_sum_by_year.index.max())

ax.tick_params(labeltop=True, labelright=True, labelsize=12)
```

## 10. Group North American, Japan, European & Other region sales data as per Genre and plot it (It should show highest of region according to game category, ex- NA_Sales is highest in Adventure category , plot kind should be bar)

```python
video_game_sales_by_genre = (
    video_game_sales_df
    .drop(columns=['year_of_release', 'games_sold_global',])
    .groupby('genre')
    .sum(numeric_only=True)
    )
video_game_sales_by_genre.sort_values('games_sold_na', ascending=False).head()
```

```python
ax = video_game_sales_by_genre.plot.bar(figsize=(20,8), rot=1, grid=True)

plt.title('total_video_game_sales_by_genre_and_region'.replace('_', ' ').title(), pad=10, fontsize=18)
plt.legend(fontsize = 12)

plt.xlabel("Genre", labelpad=10, fontsize=15)
plt.ylabel("Number of Games Sold", labelpad=10, fontsize=15)

ax.yaxis.set_major_locator(ticker.MultipleLocator(100000000))
ax.yaxis.set_major_formatter(formatter)

ax.tick_params(labelright=True, labelsize=12)
```

## 11. Group North American, Japan, European & Other region sales data as per Platform and plot it (Should show sales as per "Platform genre" , plot kind should be bar)

```python
video_game_sales_by_platform = (
    video_game_sales_df
    .drop(columns=['year_of_release', 'games_sold_global',])
    .groupby('platform')
    .sum(numeric_only=True)
)
video_game_sales_by_platform.sort_values('games_sold_na', ascending=False).head()
```

```python
ax = video_game_sales_by_platform.plot.bar(figsize=(20,8), rot=1, grid=True)

plt.title('total_video_game_sales_by_platform_and_region'.replace('_', ' ').title(), pad=10, fontsize=18)
plt.legend(fontsize = 12)

plt.xlabel("Platform", labelpad=10, fontsize=15)
plt.ylabel("Number of Games Sold", labelpad=10, fontsize=15)

ax.yaxis.set_major_locator(ticker.MultipleLocator(50000000))
ax.yaxis.set_major_formatter(formatter)

ax.tick_params(labelright=True, labelsize=12)
```

## 12. Plot the Sales in all regions as per Platform , you need to create a boxplot for that which will show the distribution of sales as per platform

```python
video_game_global_sales_by_platform = video_game_sales_df[['games_sold_global', 'platform']]
video_game_sales_group_by_platform_median = video_game_global_sales_by_platform.groupby('platform').median(numeric_only=True)
video_game_sales_group_by_platform_median_ordered = video_game_sales_group_by_platform_median.sort_values('games_sold_global')
```

```python
ax = sns.boxplot(data=video_game_global_sales_by_platform, x='platform', y='games_sold_global',
                 showfliers=False, order=video_game_sales_group_by_platform_median_ordered.index)

plt.title('distribution_of_video_game_sales_by_platform'.replace('_', ' ').title(), pad=10, fontsize=18)

plt.xlabel("Platform", labelpad=10, fontsize=15)
plt.ylabel("Number of Games Sold", labelpad=10, fontsize=15)

ax.yaxis.set_major_formatter(dec_formatter)
ax.tick_params(labelright=True, labelsize=12)
```

## 13. Find out which Genre games are sold in highest numbers (create a countplot of seaborn which could do that)

```python
ax = sns.countplot(data=video_game_sales_df[['genre']], x='genre', 
                   order=video_game_sales_df['genre'].value_counts(ascending=True).index)

sns.set(rc={'figure.figsize':(20, 8)})

plt.title('number of distinct (platform and year independent) video games sold per genre'.title(), pad=10, fontsize=18)

plt.xlabel("Genre", labelpad=10, fontsize=15)
plt.ylabel("Count of Video Games", labelpad=10, fontsize=15)

ax.tick_params(labelright=True, labelsize=12)
```

## 14. Try to find out global sales as per Genre, (create a barplot of seaborn for that)

```python
video_game_global_sales_by_genre = video_game_sales_df[['games_sold_global', 'genre']]

video_game_total_sales_by_genre = video_game_global_sales_by_genre.groupby('genre').sum().reset_index()
video_game_total_sales_by_genre_order = video_game_total_sales_by_genre.sort_values('games_sold_global')
video_game_total_sales_by_genre_order.head()
```

```python
ax = sns.barplot(data=video_game_global_sales_by_genre, x='genre', y='games_sold_global',
                 estimator='sum', order=video_game_total_sales_by_genre_order['genre'])

sns.set(rc={'figure.figsize':(20, 8)})

plt.title('total_video_game_sales_per_genre'.replace('_', ' ').title(), pad=10, fontsize=18)

plt.xlabel("Genre", labelpad=10, fontsize=15)
plt.ylabel("Number of Games Sold", labelpad=10, fontsize=15)

ax.yaxis.set_major_formatter(formatter)

ax.tick_params(labelright=True, labelsize=12)
```
